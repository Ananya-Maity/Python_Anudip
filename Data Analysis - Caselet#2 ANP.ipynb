```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path of XLSX file
file_path = 'Data Analysis - Caselet#2.xlsx'

# Read the Excel file and extract the 'DifferentFormat' sheet
df = pd.read_excel(file_path, sheet_name='DifferentFormat')

# Set up custom colors
boxplot_colors = dict(boxes='blue', whiskers='green', medians='red', caps='gray')

# Create a boxplot for the 'TPT' column, grouped by 'Team Member'
df.boxplot(column=['TPT'], by=['Team Member'], color=boxplot_colors, patch_artist=True)

# Customize the plot
plt.title("Boxplot by Team Members")
plt.xticks(rotation=90)
plt.xlabel("Team Members")
plt.ylabel("TPT")

# Show the plot
plt.show()
```
