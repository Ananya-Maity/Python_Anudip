```python
import numpy as np
from scipy.sparse import csr_matrix
```

```python
# Create a 2-D array
arr = np.array([[1,0,0],[7,3,0,],[0,5,2]])
print('Original matric:\n',arr)
```

```python
# Converting to sparse matrix
sm = csr_matrix(arr)
print(sm)
```

```python
print(sm.data)
```

```python
from scipy.sparse import csc_matrix
smc= csc_matrix(arr, shape=(3,3))
print(smc)
```
